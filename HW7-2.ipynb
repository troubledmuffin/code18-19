```python
import urllib.request
import networkx as nx
import pandas as pd
import gensim
from gensim.models import word2vec
import logging
from tqdm import tqdm
import os
a = []
c = [] 
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = gensim.models.KeyedVectors.load_word2vec_format("ruscorpora_mystem_cbow_300_2_2015.bin.gz", binary=True)
words = ["желтый_A", "красный_A", "зеленый_A", "фиолетовый_A"]
G = nx.Graph() # пустой граф
G.add_nodes_from(words)
for word in words:
    # есть ли слово в модели? 
    if word in model:
        for i in tqdm(model.most_similar(positive=[word], topn=100)):
            # слово + коэффициент косинусной близости
            if i[0].endswith('_A') and i[1] >= 0.5:
                a.append(i[0])
                G.add_nodes_from(a)
                for l in a:
                    G.add_edge(word, l)
                    for i in tqdm(model.most_similar(positive=[l], topn=100)):
                        if i[0].endswith('_A') and i[1] >= 0.5:
                            c.append(i[0])
                            G.add_nodes_from(c)
                            for t in c:
                                G.add_edge(l, t)
                                
        print('\n')
    else:
        # Увы!
        print('Увы, слова "%s" нет в модели!' % word)
print('узлы', G.nodes())
print('рёбра', G.edges())
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='green', node_size=20) 
nx.draw_networkx_edges(G, pos, edge_color='blue')
nx.draw_networkx_labels(G, pos, font_size=12, font_family='Arial')
plt.axis('off') 
plt.show()
print(nx.radius(G))
print(nx.diameter(G))
print(nx.degree_pearson_correlation_coefficient(G))
print(nx.density(G))
print(nx.average_clustering(G))
print(nx.transitivity(G))
deg = nx.degree_centrality(G)
deg2 = sorted(deg, key=deg.get, reverse=True)
print(deg2[0])
eig = nx.eigenvector_centrality(G)
eig1 = sorted(eig, key=eig.get, reverse=True)
print(eig1[0])
bet = nx.betweenness_centrality(G)
bet2 = sorted(bet, key=bet.get, reverse=True)
print(bet2[0])
clo = nx.closeness_centrality(G)
clo2 = sorted(clo, key=clo.get, reverse=True)
print(clo2[0])    

```
